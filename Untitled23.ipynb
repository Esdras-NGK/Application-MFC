```python
import os

def charger_corpus_bbc(chemin_racine):
    """
    Parcourt les dossiers (business, tech, etc.) et charge les textes.
    """
    corpus = []
    # Liste des thématiques ciblées
    categories = ['business', 'entertainment', 'tech']

    for categorie in categories:
        chemin_cat = os.path.join(chemin_racine, categorie)
        if os.path.exists(chemin_cat):
            print(f"Chargement de la catégorie : {categorie}")
            for fichier in os.listdir(chemin_cat):
                if fichier.endswith(".txt"):
                    with open(os.path.join(chemin_cat, fichier), 'r', encoding='latin-1') as f:
                        texte = f.read()
                        corpus.append({
                            'texte': texte,
                            'categorie': categorie,
                            'nom_fichier': fichier
                        })
    print(f"Total de documents chargés : {len(corpus)}")
    return corpus

# Exemple d'utilisation (remplacez par votre chemin réel)
# bbc_data = charger_corpus_bbc('path/to/bbc-fulltext')
```
